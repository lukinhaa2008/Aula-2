```python
nome = input('Escreva seu primeiro nome:')
ano_de_nascimento = int(input('Digite aqui seu ano de nascimento: '))

n1 = float(input('Valor da nota 1: '))
n2 = float(input('Valor da nota 2: '))
n3 = float(input('Valor da nota 3: '))
n4 = float(input('Valor da nota 4: '))

max(n1, n2, n3, n4)
min(n1, n2, n3, n4)

media_das_notas = (n1 + n2 + n3 + n4)/4

print(f'Nome do aluno: {nome}')
print(f'Ano de nascimento: {ano_de_nascimento}')
print(f'Média das notas: {media_das_notas}')
print(f'Maior nota: {max(n1, n2, n3, n4)}')
print(f'Menor nota: {min(n1, n2, n3, n4)}')
```

```python
funcionarios_operador = 42
funcionarios_lider = 9
funcionarios_gerente = 3
salario_operador = 3000
salario_lider = 4500
salario_gerente = 6000

media_ponderada = (funcionarios_operador * salario_operador + funcionarios_lider * salario_lider + funcionarios_gerente * salario_gerente) / (funcionarios_operador + funcionarios_lider + funcionarios_gerente)

print(f'Total de funcionarios:{funcionarios_operador + funcionarios_lider + funcionarios_gerente}')
print(f'A diferença entre o salario mais baixo e o mais alto:{salario_gerente - salario_operador}')
print(f'A média ponderada da faixa salarial é:{round(media_ponderada)}')
```

5) Escreva um programa que peça ao usuário para inserir as notas de três provas e os pesos correspondentes a cada uma delas. O programa deve calcular e exibir a média ponderada dessas notas.

peso1 = 2

peso2 = 3

peso3 = 5

image.png

```python
nota1 = float(input('Nota 1: '))
nota2 = float(input('Nota 2: '))
nota3 = float(input('Nota 3: '))

peso1 = float(input('Peso 1: '))
peso2 = float(input('Peso 2: '))
peso3 = float(input('Peso 3: '))

media_ponderada = (nota1 * peso1 + nota2 * peso2 + nota3 * peso3) / (peso1 + peso2 + peso3)
print(f'Média ponderada:{round(media_ponderada)}')
```
